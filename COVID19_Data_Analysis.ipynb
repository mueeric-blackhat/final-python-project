```python
# Step 2: Load the dataset
import pandas as pd

# Load data using pandas
df = pd.read_csv("owid-covid-data.csv")

# Preview first 5 rows
print("Preview of the dataset:")
display(df.head())

# Check all column names
print("\nColumn names:")
print(df.columns)

# Check for missing values
print("\nMissing values per column:")
print(df.isnull().sum().sort_values(ascending=False))

# Data types and basic info
print("\nDataset Info:")
df.info()
```

```python
# Step 3: Data Cleaning

# Filter countries of interest
countries = ['Kenya', 'United States', 'India']
df_countries = df[df['location'].isin(countries)].copy()

# Drop rows with missing dates or total_cases
df_countries = df_countries.dropna(subset=['date', 'total_cases'])

# Convert date to datetime
df_countries['date'] = pd.to_datetime(df_countries['date'])

# Fill missing numeric values with forward fill (per country)
numeric_cols = ['total_cases', 'new_cases', 'total_deaths', 'new_deaths', 'total_vaccinations']
df_countries[numeric_cols] = df_countries.groupby('location')[numeric_cols].transform(lambda x: x.fillna(method='ffill'))

# Preview cleaned data
print("Cleaned dataset preview:")
display(df_countries.head())
```

```python
# Step 4: Data Analysis

# Group by country to get peak values
peak_stats = df_countries.groupby('location').agg({
    'total_cases': 'max',
    'new_cases': 'max',
    'total_deaths': 'max',
    'new_deaths': 'max',
    'total_vaccinations': 'max'
}).rename(columns={
    'total_cases': 'Peak Total Cases',
    'new_cases': 'Peak New Cases (Daily)',
    'total_deaths': 'Peak Total Deaths',
    'new_deaths': 'Peak New Deaths (Daily)',
    'total_vaccinations': 'Total Vaccinations'
})

# Average daily new cases and deaths
avg_stats = df_countries.groupby('location').agg({
    'new_cases': 'mean',
    'new_deaths': 'mean'
}).rename(columns={
    'new_cases': 'Avg Daily New Cases',
    'new_deaths': 'Avg Daily New Deaths'
})

# Merge the two tables
summary_stats = peak_stats.join(avg_stats)

# Round for neatness
summary_stats = summary_stats.round(2)

# Display the result
print("COVID-19 Summary Statistics per Country:")
display(summary_stats)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style
sns.set(style="darkgrid")
plt.figure(figsize=(12, 6))

# Global trend: new cases and new deaths over time
global_trend = df.groupby('date')[['new_cases', 'new_deaths']].sum().rolling(window=7).mean()

global_trend.plot(figsize=(14,6))
plt.title("Global COVID-19 Trends: 7-Day Rolling Average")
plt.xlabel("Date")
plt.ylabel("Count")
plt.legend(["New Cases", "New Deaths"])
plt.show()
```

```python
# Top 10 countries by total cases
top_cases = df_countries.groupby('location')['total_cases'].max().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_cases.values, y=top_cases.index, palette="Reds_r")
plt.title("Top 10 Countries by Total COVID-19 Cases")
plt.xlabel("Total Cases")
plt.ylabel("Country")
plt.show()
```

```python
# Top 10 countries by total deaths
top_deaths = df_countries.groupby('location')['total_deaths'].max().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_deaths.values, y=top_deaths.index, palette="Greys_r")
plt.title("Top 10 Countries by Total COVID-19 Deaths")
plt.xlabel("Total Deaths")
plt.ylabel("Country")
plt.show()
```

```python
# Vaccination progress for top 10 countries
top_vax = df_countries.groupby('location')['total_vaccinations'].max().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_vax.values, y=top_vax.index, palette="Greens_r")
plt.title("Top 10 Countries by Total Vaccinations")
plt.xlabel("Total Vaccinations")
plt.ylabel("Country")
plt.show()
```
